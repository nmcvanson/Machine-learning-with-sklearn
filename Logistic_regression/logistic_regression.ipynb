```python
import pandas as pd
```

```python
dataset = pd.read_csv("purchase_records.csv")
dataset
```

```python
from sklearn.preprocessing import  LabelEncoder
labelEncoder_gender = LabelEncoder()
Gender_Number = labelEncoder_gender.fit_transform(dataset["Gender"])
dataset.insert(1,'Gender_Number', Gender_Number)
dataset
```

```python
#X = dataset[['Gender_Number', 'Age', 'Salary']]
#y = dataset[['Purchase Iphone']]
X = dataset.iloc[:, 1:4]
y = dataset.iloc[:, 4]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0, solver="liblinear")
classifier.fit(X_train, y_train)
```

```python
y_test
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, y_pred) 
print(cm)
accuracy = metrics.accuracy_score(y_test, y_pred) 
print("Accuracy score:",accuracy)
precision = metrics.precision_score(y_test, y_pred) 
print("Precision score:",precision)
recall = metrics.recall_score(y_test, y_pred) 
print("Recall score:",recall)
```
